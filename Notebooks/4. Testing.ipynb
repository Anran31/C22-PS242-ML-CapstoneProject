# Load Testing Data

# Load Trained Model

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
model = load_model(root + 'Saved/model_lstm.hdf5')
```

# Testing

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from sklearn import preprocessing
```

```python
max_length = 120
trunc_type='post'
padding_type='post'
```

```python
sentences = [str(text) for text in df.text]
sequences = tokenizer.texts_to_sequences(sentences)
X_test = pad_sequences(sequences, maxlen=max_length, padding=padding_type, truncating=trunc_type)
```

```python
label_encoder = preprocessing.LabelEncoder()
df['label']= label_encoder.fit_transform(df['label'])  
y_test = np.array(df['label'])
y_test = to_categorical(y_test)
```

```python
label_encoder.classes_
```

# Predict

```python
pred = model.predict(X_test)
y_pred = np.argmax(pred, axis=-1)
```

```python
y_test = [np.argmax(y, axis=None, out=None) for y in y_test]
```

```python
from sklearn.metrics import plot_confusion_matrix, confusion_matrix, classification_report, f1_score
```

```python
print(classification_report(y_test, y_pred))
```
