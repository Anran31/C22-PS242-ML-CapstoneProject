```python
import pandas as pd
import numpy as np
```

```python
datasets = pd.DataFrame(columns=['text', 'label'])
```

# Load Hotel Dataset

```python
hotel_dataset_path = 'Datasets/classes/hotel/hotel_review.csv'
```

```python
df = pd.read_csv(root + hotel_dataset_path, nrows=1000, encoding='ISO-8859-1')
df.head()
```

```python
for text in df.Review:
  datasets.loc[len(datasets.index)] = [text, 'hotel']
```

# Load Restaurant Dataset

```python
restaurant_dataset_path = 'Datasets/classes/restaurant/Restaurant_Reviews.tsv'
```

```python
df = pd.read_csv(root + restaurant_dataset_path, sep='\t')
df.head()
```

```python
for text in df.Review:
  if text.strip() == '':
    print('ada')
    continue
  datasets.loc[len(datasets.index)] = [text, 'restaurant']
```

# Load Museum Dataset

```python
museum_dataset_path = 'Datasets/classes/museum/tripadvisor_merged.csv'
```

```python
df = pd.read_csv(root + museum_dataset_path)
df = df.dropna(subset=['Description'])
df.head()
```

```python
for text in df.Description:
  datasets.loc[len(datasets.index)] = [text, 'museum']
```
